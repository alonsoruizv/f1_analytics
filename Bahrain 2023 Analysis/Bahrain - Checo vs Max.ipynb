```python
import fastf1
from fastf1 import plotting
from matplotlib import pyplot as plt
fastf1.Cache.enable_cache('/Users/alonsoruizvelascoangeles/Repos/f1_analytics/Throttle Analysis/cache')  
```

```python
race = fastf1.get_session(2023, 'Bahrain', 'R')
race.load()

per = race.laps.pick_driver('PER')
ver = race.laps.pick_driver('VER')
```

```python
plotting.setup_mpl()

fig, ax = plt.subplots()
ax.plot(per['LapNumber'], per['LapTime'], color='red')
ax.plot(ver['LapNumber'], ver['LapTime'], color='cyan')
ax.set_title("PER vs VER")
ax.set_xlabel("Lap Number")
ax.set_ylabel("Lap Time")
ax.legend()
plt.show()
```
